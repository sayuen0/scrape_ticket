```python
import requests
from bs4 import BeautifulSoup
import re


#{YYYY-mm-dd} ãªæ—¥ä»˜ã‘ã‚’ä½œæˆ

def formatted_today():
    import datetime
    now = datetime.datetime.now()
    today = "{0:%Y-%m-%d}".format(now)
    return today

#åˆå›ã®ã¿
target_url = "https://www.yahoo.co.jp/"
#Requestsã‚’ä½¿ã£ã¦ã€webã‹ã‚‰å–å¾—
r = requests.get(target_url)
#è¦ç´ ã‚’æŠ½å‡º
soup = BeautifulSoup(r.text, 'html.parser')


today = formatted_today()

import os 
os.mkdir(today)

#HTMLãƒ•ã‚¡ã‚¤ãƒ«ã¨ã—ã¦ä¿å­˜ã—ãŸã„å ´åˆã¯ãƒ•ã‚¡ã‚¤ãƒ«ã‚ªãƒ¼ãƒ—ãƒ³ã—ã¦ä¿å­˜
with open(today+"/"+'yahoo.html', mode='w', encoding = 'utf-8') as fw:
    fw.write(soup.prettify())

#soup.find_allã‚’ç”¨ã„ã¦ãƒªãƒ³ã‚¯å…ˆãŒã€Œnews.yahoo.co.jp/pickupã€ã®é …ç›®ã‚’å…¨ã¦å–å¾—
elems = soup.find_all(href=re.compile("news.yahoo.co.jp/pickup"))
for e in elems:
    print(e.getText())
    
```

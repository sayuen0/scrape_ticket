```python
# ランダムサンプリング


df = pd.read_csv("../分析対象データ群/only-jannies.csv",index_col=0)
df_enka = pd.read_csv("../CSV/演歌/演歌.csv",index_col=0)
```

```python
#500回ランダムサンプリング 
#それぞれを演歌と結合して保存
for i in range(1,501):
    d_sampled = df.sample(n=1600)
    d_sampled.to_csv("../分析対象データ群/ジャニーズグループのみランダムサンプリングCSV/ジャニーズグループのみランダムサンプリング"+str(i).zfill(3)+"回目.csv") 
    d_concated = pd.concat([d_sampled,df_enka])
    d_concated.to_csv("../分析対象データ群/ジャニーズをランダムサンプルのち演歌と結合CSV/ジャニーズをランダムサンプルのち演歌と結合"+str(i).zfill(3)+"回目.csv")    
```

```python
# 平均化
df = pd.read_csv("../分析対象データ群/only-jannies.csv")
df
```

```python
# 平均を取得
d_grouped = df.groupby(["name","date","collected_at","left_days","artistname"],as_index=False)["price_per_unit"].mean()
# d_grouped.to_csv("../分析対象データ群/収集日公演グループ済みジャニーズ群.csv")
# 分析用列追加
# アーティストダミー
d_grouped["is_jannies"] = 1
# 法施行ダミーと交差項
d_grouped["after"] = (pd.to_datetime(d_grouped["collected_at"],format="%Y-%m-%d") >= pd.to_datetime("2019-06-14",format="%Y-%m-%d")) *1
d_grouped["after_and_is_jannies"] = d_grouped["after"] * d_grouped["is_jannies"] 
d_grouped
# d_grouped.to_csv("../分析対象データ群/平均集約/集約済みジャニーズ.csv")
```

```python

# 列名変更 単位価格→平均単位価格に
# d_grouped.rename(columns={"price_per_unit":"average_price_per_unit"})
# type(d_grouped)


def add_age(df):
    ages = []
    for index, row in df.iterrows():
        if row["artistname"] =="ジャニーズJr":
#             row["age"] = 22.3
            ages.append(22.3)
        elif row["artistname"]== "セクシーゾーン":
#             row["age"] = 22.2
            ages.append(22.2)
        elif row["artistname"] == "キスマイフットツー":
            ages.append(30.5)
#             row["age"]= 30.5
        elif row["artistname"]== "NEWS":
            ages.append(32.75)
#             row["age"] = 32.75
    df["age"] = ages
    return df
#     else:
#         df["age"] = None
d_grouped = add_age(d_grouped)
d_grouped
# d_grouped.to_csv("../分析対象データ群/収集日公演グループ済みジャニーズ群.csv",index=0)

d_grouped.to_csv("../分析対象データ群/平均集約/集約済みジャニーズ.csv")
```

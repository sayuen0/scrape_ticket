```python
import pandas as pd
import numpy as np
import re
from bs4 import BeautifulSoup
import datetime

```

```python
# ローカルファイルを取得
# ジャニーズJr/2019-05-16をコピーして用意
soup = BeautifulSoup(open("./jannies.html"),"html.parser")
soup
```

```python
# 全角→半角関数
def get_hankaku_from_zenkaku(text):
    return text.translate(str.maketrans({chr(0xFF01 + i): chr(0x21 + i) for i in range(94)}))

# 公演名取得関数
def get_kouen_name(selected_soup=""):
    kouen_name_list =  list(\
                 map(\
                     lambda x:
                     get_hankaku_from_zenkaku(
                         x.string.strip().replace("\n","").replace("\u3000","")
                     ),
                     selected_soup))
    return kouen_name_list


# 日付フォーマット関数
def get_formatted_date(datestr):
    return datetime.datetime.strptime(datestr,"%Y/%m/%d")

# スープからタグ除去関数
# def remove_soup_tag(soup):
#     p = re.compile(r"<[^>]*?>")
#     return list( map(lambda x:p.sub("",str(x)).replace(" ",""), soup)
            

# def remove_single_str_tag(str):
#     p = re.compile(r"<[^>]*?>")
#     return p.sub("", str).replace(" ", "")
                    

```

```python
dataframe = pd.DataFrame()
# 公演名を取得
# str型
#前後空白除去 strip()
# 置き換え replace("\n", "")


kouen_name_list= list(\
                 map(\
                     lambda x:x.string.strip().replace("\n",""),
                     soup.select(".list-ticket-name>p")))


dataframe["name"] = kouen_name_list
dataframe
```

```python
#公演日時を取得
# Timestamp型
#スープのうちタグに相当するもの <p>や</p> を正規表現で削除

p = re.compile(r"<[^>]*?>")
kouen_data_list = list( map(\
                     lambda x:
                            datetime.datetime.strptime( #str→datetime
                                "20"+ #元の文字が19/05/17で年数に不備が生じるので、先頭に20をつける
                                p.sub("",str(x)) #タグ除去
                                .replace(" ","") #空白除去
                                .replace("\n", "")[:8],"%Y/%m/%d"),#改行除去、曜日を削除、フォーマット変換
                     soup.select(".list-ticket-date>span")))


# type(kouen_data_list[0])
dataframe["date"] = kouen_data_list
dataframe
```

```python
#チケットデータ収集日を取得。
#収集フォルダに日時を記載している(今回は2019/05/16)
# チケットデータ収集日と公演日時
datestr = "2019/05/16"


dataframe["collected_at"] = datetime.datetime.strptime(datestr,"%Y/%m/%d")
dataframe
```

```python
#公演までの残り日数を取得
# 予想：公演が早くなるにつれ
# 公演日時 - データ収集日
dataframe["until"] = (dataframe["date"] - dataframe["collected_at"]).dt.days  #日時差をint変換
print(type(dataframe["until"][0]))
dataframe

# アーティストグループダミー
# ジャニーズなら1, 演歌なら0とする
# これはプログラムを走らせる際に指定
dataframe["isJannies"] = 1



# 購入可能ダミー　買えるなら1, 買えないなら0
# 買えるとき →　.list-ticket-order内がformになっているとき

list_ticket_order = soup.select(".list-ticket-order")

available = []
for i, order in enumerate(list_ticket_order):
    if(order.find("form")):
        available.append(1)
    else:
        available.append(0)
        
dataframe["available"] = available
dataframe

```

```python
#取引あたり価格データおよび、単位あたり価格データを取得
#何枚か連番になっているので、連番枚数で除して単位あたり価格を取得

# 取引あたり価格
deal_price_list = list(map(lambda x:
                      int(re.findall(r"￥\d+,\d+",
                                 str(x))[0][1:]
                                 .replace(",", "")),
                      soup.select(".list-ticket-price")))
dataframe["price_per_deal"] = deal_price_list
dataframe

#連番数

price_list = soup.select(".list-ticket-price")
unit_list = []
for prc in price_list:
    if "ticket-sellunit" in str(prc):
        unit_list.append(
            int(prc.select(".ticket-sellunit")[0].get_text()))
    else:
        unit_list.append(1)
        
dataframe["unit"] = unit_list
dataframe["price_per_unit"] = dataframe["price_per_deal"]/ dataframe["unit"]
dataframe
```

```python
#　チケット転売施工ダミーを儲ける 施工日は6月14日
# 収集日がどうかで決める
#  after_constructed   1なら施工後、0なら施工前
# *1することでTrue ,False→ 1, 0

def is_after_constructed(dataframe):
    return (dataframe["collected_at"] >= "2019-06-14")*1

dataframe["after_constructed"] =  is_after_constructed(dataframe)
dataframe
```

```python
# CSV書き出し

dataframe.to_csv("jannies-ticket-data-sample.csv",index=0)
df = pd.read_csv ("jannies-ticket-data-sample.csv")
df
```

```python
# 結局取得するコラムは順に
# name
# date
# collected_at
# unitl
# isJannies
# available
# price_per_deal
# unit
# price_per_unit
# after_constructed
```

```python
# 演歌の方に着手
soup2 = BeautifulSoup(open("./enka.html"),"html.parser")
soup2

dataframe2 = pd.DataFrame()

# 公演名: name
names = get_kouen_name(soup2.select(".Product__titleLink"))
# len(names)
dataframe2["name"] = names
dataframe2

# データ収集日: collected_at,after_constructed 
dataframe2["collected_at"] = get_formatted_date("2019/05/16")
dataframe2["after_constructed"]  = is_after_constructed(dataframe2)

# 価格について:price_per_unit/price_per_deal
# 即決価格はないときもあればあるときもある
# 現在価格は必ずある
# 即決価格があれば即決価格を、なければ現在価格をprice_per_dealとして採用することにする
# 現在価格のセレクタは.Product__priceValue.u-textRed 即決価格は.Product__priceValue

price_list = []
price_soup =soup2.select(".Product__priceInfo")
for s in price_soup:
    if(s.select(".Product__priceValue:not(.u-textRed)")):
        s = s.select(".Product__priceValue:not(.u-textRed)")[0].text.replace(" ", "").replace("円", "").replace(",", "")
        s = int(s)
        price_list.append(s)
    else:
        s = s.select(".Product__priceValue.u-textRed")[0].text.replace(" ", "").replace("円", "").replace(",", "")
        s =int(s)
        price_list.append(s)
dataframe2["price_per_deal"] = price_list
dataframe2

dataframe2["isJannies"]= 0
dataframe2

# availableについて
# (未確認)出品終了していない限り手に入るものとする → 1

dataframe2["available"] = 1
```

```python
# 連番について/unit
# nameに「〇〇枚」の表現があれば連番枚数として採用。なければ1枚

unit_list = []
r_pattern = re.compile(r"\d+枚")
df  = pd.DataFrame()

for i in dataframe2["name"]:
    if(r_pattern.search(i)):
        found = r_pattern.search(i)[0]
        num = int(re.sub("\\D", "", found))
        unit_list.append(num)
    else:
        unit_list.append(1)
# r_pattern.search(text)
# print(text)
dataframe2["unit"] = unit_list
dataframe2


# 単位あたり価格/price_per_unit

dataframe2["price_per_unit"] = dataframe2["price_per_deal"]/ dataframe2["unit"]
dataframe2
```

```python
# dateについて
# nameのなかに日付表現があればそれを採用
# なければ取引終了日とする
# 残り日数は.Product__timeに入っている
# 

date_pattern = re.compile(r"(\d{1,4}\/\d{1,2}\/\d{1,2}|\d月\d+日|\d{1,3}\/\d+)")
nokorizikan = soup2.select(".Product__time") #残り時間
collected_at = datetime.date(2019,5,17) #収集日
print(collected_at)

date_list = []
for i , data in enumerate(dataframe2["name"]):
#     商品名に日付表現があれば
    if(date_pattern.search(data)):
#         見つかったやつから生成した日付がdate
        found = date_pattern.search(data)[0]
        found = found.replace("月", "/").replace("日", "")
        nums = [int(s) for s in found.split("/")]
        if(len(nums) <= 2): #年がなかったら
            this_year = datetime.datetime.today().year
            nums.insert(0,this_year)
        d = datetime.date(*nums)
        date_list.append(d)
#       商品名に日付表現がなければ
    else:
#         同じアイテムの残り日数/時間を収集日に加算したものがdate
        nokori =nokorizikan[i].text.strip()
#         日を含んでいれば収集日にその日づけを加算
        if "日" in nokori:
            d = collected_at + datetime.timedelta(days=int(re.sub("\\D", "", nokori)))
            date_list.append(d)
#             含んでいなけれその日(時間は更新する必要ないので)
        else:
            d = collected_at 
            date_list.append(d)

dataframe2["date"] = date_list
# datetime.date型で収めてしまったのでtimestampに変換
dataframe2["date"] = pd.to_datetime(dataframe2["date"])

# dataframe2

dataframe2["until"] = (dataframe2["date"] - dataframe2["collected_at"]).dt.days  #日時差をint変換
dataframe2
```

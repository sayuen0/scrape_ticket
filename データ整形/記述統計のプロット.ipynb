```python
df = pd.read_csv("../only-jannies.csv",index_col=0)
```

```python
# 記述統計のプロット
# TODO: アイドルグループの日毎の記述統計量のプロット
# 平均と標準偏差と四分位数があれば良いと思う

def extract_statistic_and_change_name(group,statistic_str):
    s = pd.Series()
    if statistic_str =="mean":
        s = group.mean()
    elif statistic_str =="std":
        s = group.std()
    elif statistic_str =="max":
        s = group.max()
    elif statistic_str =="min":
        s = group.min()
    s.name = f"{statistic_str}_price"
    return s
    
d_grouped = df.groupby("collected_at")["price_per_unit"]
d_mean = extract_statistic_and_change_name(d_grouped, "mean")
d_std  = extract_statistic_and_change_name(d_grouped, "std" )
d_max  = extract_statistic_and_change_name(d_grouped, "max" )
d_min  = extract_statistic_and_change_name(d_grouped, "min" )



jannies_date_grouped= pd.concat([d_mean,d_std,d_max,d_min],axis=1)
# jannies_date_grouped.shape #(83, 4)
jannies_date_grouped.to_csv("../分析対象データ群/ジャニーズグループ日付ごと記述統計.csv")


```
